```python
## Finding the Leading Cause of Deaths in all States and all years

## Drop Uneeded Columns
df = df[['Cause Name', 'Deaths']]

## Grouping by Cause
grouped = df.groupby(['Cause Name'], sort=True).sum()

## Sort Descending by Death
grouped.sort_values(by=['Deaths'], ascending=False)
```

```python
## Finding the Leading Cause of Death in DC

df = pd.read_csv(filename)

## Remove Extra Columns
df = df[['Cause Name', 'State', 'Deaths']]

## Grouping By State and Cause
grouped = df.groupby(by=['State', 'Cause Name']).sum()

## Select out DC and sort by descending
grouped.loc['District of Columbia'].sort_values(by=['Deaths'], ascending=False)
```

```python
## What are the top 5 states that where Heart Disease is the most common

df = pd.read_csv(filename)

## Remove Extra Columns
df = df[['Cause Name', 'State', 'Deaths']]

## Grouping By State and Cause
grouped = df.groupby(by=['Cause Name', 'State']).sum()

grouped.loc['Heart disease'].sort_values(by=['Deaths'], ascending=False)
```

```python
## Read in Cencus Data
dfCEN = pd.read_excel(filename2)

## Change Column names
dfCEN = dfCEN.rename(columns={ dfCEN.columns[0]: "State" })
```

```python
## Grab Heart Disease Rates for USA
HDDeaths = grouped.loc['Heart disease'].sort_values(by=['Deaths'], ascending=False)

## Remove Period in front of State Name
dfCEN['State'] = [x[1:] for x in dfCEN.State]

## Merge together the two dataframes
Combo = HDDeaths.merge(dfCEN, on='State')

## Check Dataframes
Combo
```

```python
## Remove Extra Columns
Combo = Combo[['State', 'Deaths', 2013]]

Combo.head()
```

```python
## Create coloum the Average Death per Person
Combo['Avg Death'] = Combo['Deaths']/Combo[2013]
```

Finding the Top 5 States with the highest Death from Heart Disease per Person

```python
Combo.sort_values(by=['Avg Death'], ascending=False).head()
```

Finding the Bottom 5 States with the highest Death from Heart Disease per Person

```python
Combo.sort_values(by=['Avg Death'], ascending=False).tail()
```
